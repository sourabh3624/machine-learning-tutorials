```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
##simple i.e:

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
## plotting using matplotlib
##plt scatter
plt.scatter(x,y,c='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('graph in 2D')
plt.savefig('2d.png')
```

```python
y=x*x
```

```python
plt.plot(x,y,"r*",linestyle='dashed',linewidth=3,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('graph in 2D')
plt.savefig('2d1.png')
```

```python
# creating subplot:
plt.subplot(2,2,1)
plt.plot(x,y,'g')
plt.subplot(2,2,2)
plt.plot(x,y,'r')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
x=np.arange(1,11)
y=3 * x + 5
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('graph in 2D')
plt.plot(x,y)
```

```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('sine wave form')
plt.plot(x,y)
```

```python
x=np.arange(0,4*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1)
plt.title('sine')
plt.plot(x,y_sin,'g--')
plt.subplot(2,1,2)
plt.title('cos')
plt.plot(x,y_cos,'r--')
plt.show()
```

```python
#Bar plot:
x=[23,45,23]
y=[3,4,2]

x2=[23,435,23]
y2=[33,24,42]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('bar in 2D')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
#histogram:
a=np.array([12,32,32,34,564,54,35,32,4,24])
plt.hist(a,bins=30)
plt.title("histogram")
plt.show()
```

```python
#box plot:

data=[np.random.normal(0,std,100)for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
#pie chart:
labels='python','c++','javascript','c'
sizes=[120,40,20,20]
colors=['green','red','blue','yellow']
explode=(0.1,0,0,0)

#plot:
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)

plt.axis('equal')
plt.show()
```
